### Explorando a base de resultados e preparando os dados

```python
dados_resultados = pd.read_csv('Dados/dados_resultados.csv')
dados_resultados.head()
```

```python
dados_resultados['5-alpha_reductase_inhibitor'].unique()
```

Analisando os dados resultados podemos ver que ele se trata de um conjunto de dados em que cada linha equivale a um dos experimentos da base de experimentos. Além disso, ignorando a coluna id, as demais colunas representam um inibidor ou agonista que foi ativado ou não (ou 1 ou 0)

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
dados_resultados['ativo_moa'] = dados_resultados['n_moa'] != 0
dados_resultados.head()
```

Para facilitar a construção do modelo criamos duas novas colunas no dados de resultado. A coluna 'ativo_moa' mostra True/False pro caso de ter ativado pelo menos 1 inibidor ou agonista. Já a coluna 'n_moa' mostra a quantidade de vezes que algum inibidor ou agonista foi ativado.

```python
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

Aqui nós combinamos as colunas que criamos acima com os dados do experimento.

```python
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
dados.value_counts('tratamento')
```

Podemos ver que todos os tratamentos que estavam no grupo de controle não ativaram nenhum inibidor ou agonista. Isso está correto, já que é o grupo de refêrencia do experimento é esperado que ele não tenha ativado nada. Com isso devemos remove-lo da nossa base antes de treinarmos o nosso modelo preditivo.

```python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
dados_combinados = dados_combinados.query('tratamento == "com_droga"')
dados_combinados
```

### Treinando os modelos

Antes de treinar os modelos vamos tirar algumas colunas que não são importantes para treinar o modelo, como o 'id', 'n_moa' e a 'droga'

```python
dados_combinados = dados_combinados.drop(['id', 'n_moa', 'droga'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

O modelo de Regressão Logistica teve uma acurácia de 64%, valor abaixo do aceitável para um modelo da área de saúde.

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

O modelo Dummy chegou até a obter uma acurácia maior que a da Regressão Logística. Vamos testar outros modelos e ver como eles se comportam.

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)
modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

Utilizando uma arvóre de decisão com uma profundidade de no máximo 3 chegamos a acurácia de 66%, um valor ainda baixo para área da saúde.

```python
from sklearn import tree
fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_arvore,
              ax = ax,
              fontsize = 10, 
              rounded = True, 
              filled = True, 
              feature_names = x_treino.columns, 
              class_names = ['Não Ativado', 'Ativado'])
plt.show()
```

Aqui podemos ver gráficamente a árvore de decisão que o algoritmo criou.

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)
teste = []
treino = []
for i in range(1, 15):
    modelo_arvore = DecisionTreeClassifier(max_depth=i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
sns.lineplot(x=range(1,15), y=teste, label='teste')
sns.lineplot(x=range(1,15), y=treino, label='treino')
```

Aqui colocamos o modelo de árvore de decisão num loop para ver se aumentar a profundidade da árvore também aumenta a acurácia dela. Vemos que quanto mais profunda é a árvore mais ela conhece os dados de treino, em contrapartida, quando a apresentamos aos dados de teste o seu desempenho vai caindo. Vamos tentar com o algoritomo conhecido como RandomForest e ver como ele se sai, no caso do RandomForest temos que transformar as colunas não-númericas em colunas Dummy, que possuem ou 0 ou 1.

```python
pd.get_dummies(dados_combinados, columns=['dose', 'tempo'])
```

Nesse exemplo podemos ver o resultado da função pd.get_dummies(), onde para cada valor em 'dose' e 'tempo' ela criou uma nova coluna Dummy.

```python
from sklearn.ensemble import RandomForestClassifier

x = dados_combinados.drop(['ativo_moa', 'tratamento'], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)
modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)
```

Nenhum dos algoritmos trouxe o valor satisfatório para nós. Vamos tentar seguir o fluxograma do [Scikit-Learn](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) que pode nos auxiliar a escolher um algoritmo mais apropriado para nosso problema.

<img src="https://scikit-learn.org/stable/_static/ml_map.png">

Vamos tentar o Linear SVC

```python
from sklearn.svm import LinearSVC

x = dados_combinados.drop(['ativo_moa', 'tratamento'], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)
modelo_linearsvc = LinearSVC(max_iter=10000)
modelo_linearsvc.fit(x_treino, y_treino)
modelo_linearsvc.score(x_teste, y_teste)
```

Mesmo com o número máximo de iterações definido em 10 mil, o algoritmo de LinearSVC ainda não trouxe um resultado satisfatório.

Vimos que nenhum dos algoritmos testados foram satisfatórios. Vamos tentar balancear os dados e ver como eles se comportam já que 65% dos nossos dados ativaram algum inibidor ou agonista.

```python
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

```python
from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)

x_treino_smote, y_treino_smote = SMOTE.fit_resample(x_treino, y_treino)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()

x = dados_combinados.drop(['ativo_moa', 'tratamento'], axis=1)
x = pd.get_dummies(x, columns=['dose', 'tempo'])
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, 
                                                        test_size = 0.3,
                                                        stratify=y)

x_treino_smote, y_treino_smote = SMOTE.fit_resample(x_treino, y_treino)

modelo_linearsvc = LinearSVC(max_iter=10000)
modelo_linearsvc.fit(x_treino, y_treino)
modelo_linearsvc.score(x_teste, y_teste)
```
